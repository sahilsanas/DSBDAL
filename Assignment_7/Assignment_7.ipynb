```python
import math, re, sys, subprocess

try:
    import nltk
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "nltk"])
    import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from pathlib import Path

def read_doc(filename):
    candidates = [
        Path('/content/drive/MyDrive/Colab Notebooks/Assignment_7') / filename,
        Path.cwd() / filename,
        Path('c:/Users/Lenovo/Downloads') / filename,
    ]
    for path in candidates:
        if path.exists():
            with open(path, 'r', encoding='utf-8-sig') as file:
                return file.read()
    print(f"Warning: {filename} not found, using fallback text.")
    if filename == 'doc1.txt':
        return 'The quick brown fox jumps over the lazy dog.'
    return 'A quick fox and a smart dog became friends.'

doc = read_doc('doc1.txt')
doc2 = read_doc('doc2.txt')
```

```python
word_tokens = nltk.word_tokenize(doc.lower())
print(word_tokens)
print(len(word_tokens))
```

```python
print(stopwords.words("english"))
print("-->")
stop_words = set(nltk.corpus.stopwords.words('english'))
word_tokens = [token for token in word_tokens if token not in stop_words]
print(word_tokens)
print(len(word_tokens))
```

```python
tagged = nltk.pos_tag(word_tokens)
print(tagged)
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in word_tokens]
print(lemmatized_tokens)
```

```python
stemmer = nltk.stem.PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in word_tokens]
print(stemmed_tokens)
```

```python
def calculate_term_freq(tokens):
    tf_dict = {}
    for word in tokens:
        tf_dict[word] = tf_dict.get(word, 0) + 1

    total_tokens = len(tokens)
    if total_tokens == 0:
        return {}

    tf = {}
    for word, count in tf_dict.items():
        tf[word] = count / total_tokens
    return tf
```

```python
tf = calculate_term_freq(word_tokens)

doc2_tokens_for_tf = [
    token
    for token in nltk.word_tokenize(doc2.lower())
    if token not in stop_words
]
tf2 = calculate_term_freq(doc2_tokens_for_tf)

print("Term Frequency of Doc 1: ", tf, "\n\n")
print("Term Frequency of Doc 2: ", tf2)
```

```python
d1_tokens = [
    token
    for token in nltk.word_tokenize(doc.lower())
    if token not in stop_words
]
d2_tokens = [
    token
    for token in nltk.word_tokenize(doc2.lower())
    if token not in stop_words
]

def calculate_idf(d1_tokens, d2_tokens):
    N = 2
    all_tokens = set(d1_tokens + d2_tokens)
    idf = {}
    for token in all_tokens:
        f = 0
        if token in d1_tokens:
            f += 1
        if token in d2_tokens:
            f += 1
        idf[token] = math.log(N / f) if f > 0 else 0.0
    return idf

idf = calculate_idf(d1_tokens, d2_tokens)
```

```python
d1_repr = []
for token in d1_tokens:
    d1_repr.append((token, tf.get(token, 0.0) * idf.get(token, 0.0)))

d2_repr = []
for token in d2_tokens:
    d2_repr.append((token, tf2.get(token, 0.0) * idf.get(token, 0.0)))
```
